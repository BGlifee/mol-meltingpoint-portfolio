```python
from pathlib import Path
import numpy as np
import pandas as pd
from scipy.sparse import load_npz
import sys

# project root & import path
PROJECT_ROOT = Path.cwd().resolve().parents[0]
sys.path.append(str(PROJECT_ROOT / "src"))

from config import DATA_PROC, RESULTS_DIR, N_SPLITS, SEED
from models import get_models
from train_utils import cross_validate_models, append_results_table

# load features
X_train = load_npz(DATA_PROC / "X_train_tabred_default.npz").toarray()
y_train = np.load(DATA_PROC / "y_train_default.npy")

# run CV
models = get_models()
results = cross_validate_models(
    models, X_train, y_train,
    n_splits=N_SPLITS,
    random_state=SEED,
)

# save + display
df_results = append_results_table(
    results,
    experiment_name="baseline_tabred",
    csv_path=RESULTS_DIR / "metrics_summary.csv",
)

df_results.sort_values("mean_RMSE")

```
